```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data Science Salary 2021 to 2023.csv')
```

```python
df.head(10)
```

```python
df = df.dropna()
```

```python
df['salary_in_usd'] = pd.to_numeric(df['salary_in_usd'], errors='coerce')
```

Problem 1: Salary Trend Over Years

```python
yearly_avg_salary = df.groupby('work_year')['salary_in_usd'].mean()
plt.figure(figsize=(8, 5))
yearly_avg_salary.plot(kind='line', marker='o')
plt.title('Average Data Science Salary Trend (2021-2023)')
plt.xlabel('Year')
plt.ylabel('Average Salary (USD)')
plt.grid(True)
plt.show()
```

```python
print("Yearly Average Salaries:\n", yearly_avg_salary)
```

Problem 2: Salary by Experience Level

```python
exp_level_avg_salary = df.groupby('experience_level')['salary_in_usd'].mean().sort_values()
plt.figure(figsize=(8, 5))
exp_level_avg_salary.plot(kind='bar')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.show()
```

```python
print("Average Salary by Experience Level:\n", exp_level_avg_salary)
```

Problem 3: Top 5 Highest-Paying Job Titles

```python
job_title_avg_salary = df.groupby('job_title')['salary_in_usd'].agg(['mean', 'count'])
job_title_avg_salary = job_title_avg_salary[job_title_avg_salary['count'] >= 5]
top_5_job_titles = job_title_avg_salary['mean'].sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
top_5_job_titles.plot(kind='bar')
plt.title('Top 5 Highest-Paying Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
print("Top 5 Job Titles by Average Salary:\n", top_5_job_titles)
```

Problem 4: Top 5 Locations by Average Salary

```python
location_avg_salary = df.groupby('company_location')['salary_in_usd'].agg(['mean', 'count'])
location_avg_salary = location_avg_salary[location_avg_salary['count'] >= 5]
top_5_locations = location_avg_salary['mean'].sort_values(ascending=False).head(5)
plt.figure(figsize=(8, 5))
top_5_locations.plot(kind='bar')
plt.title('Top 5 Company Locations by Average Salary')
plt.xlabel('Company Location')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.show()
```

```python
print("Top 5 Locations by Average Salary:\n", top_5_locations)
```

Problem 5: Salary by Company Size

```python
company_size_avg_salary = df.groupby('company_size')['salary_in_usd'].mean().sort_values()
plt.figure(figsize=(8, 5))
company_size_avg_salary.plot(kind='bar')
plt.title('Average Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=0)
plt.show()
```

```python
print("Average Salary by Company Size:\n", company_size_avg_salary)
```
